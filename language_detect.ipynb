```python
page_file = pd.read_csv('/content/drive/MyDrive/FactGrid Cuneiform (AWCA)/geography/ocr-output/0_Attinger - A propos de AK «faire» (I) (ZA 95, 2005) 46-64/page.csv')
page_file.head()
```

```python
page_file["text"][0].replace("\n", " ")
```

```python
content = [re.sub("\n", " ", i) for i in page_file["text"]]
```

```python
# iso_predict = [model.predict(doc)[0].replace('__label__', '') for doc in content]
```

```python
iso_predict_0 = [model.predict(doc, k =3)[0][0][9:12] for doc in content]
iso_predict_1 = [model.predict(doc, k =3)[0][1][9:12]  for doc in content]
iso_predict_2 = [model.predict(doc, k =3)[0][2][9:12]  for doc in content]
```

```python
iso_predict_0[0]
```

```python
language_predict_0 = [languages.get(part3=i[:3]).name.lower() for i in iso_predict_0]
language_predict_1 = [languages.get(part3=i[:3]).name.lower() for i in iso_predict_1]
language_predict_2 = [languages.get(part3=i[:3]).name.lower() for i in iso_predict_2]
```

```python
import pycountry

def alpha3_to_alpha2(alpha3_code):
    country = pycountry.countries.get(alpha_3=alpha3_code)
    if country:
        return country.alpha_2.lower()
    else:
        return None
```

```python
alpha2_0 = [alpha3_to_alpha2(i) for i in iso_predict_0]
alpha2_1 = [alpha3_to_alpha2(i) for i in iso_predict_1]
alpha2_2 = [alpha3_to_alpha2(i) for i in iso_predict_2]
```

```python
page_lan = pd.DataFrame({"text": content, "pred_lan_0":alpha2_0,"lan_0":language_predict_0,"pred_lan_1":alpha2_1,"lan_1":language_predict_1,"pred_lan_2":alpha2_2,"lan_2":language_predict_2})
page_lan.head()
```

```python
page_lan_word = page_lan.assign(text=page_lan['text'].str.split('\s+')).explode('text')
page_to_word = word_file.merge(page_lan_word,  left_on = 'word', right_on= 'text').drop_duplicates(subset=['identifier']).sort_values(by = 'identifier')
page_to_word.head()
```
